```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
csv_file = '/Users/Zachary/Library/CloudStorage/OneDrive-UniversityofCambridge/Project/total_turnover_assay/TTA_result.csv'  
df = pd.read_csv(csv_file)
df.columns = ['Index', 15, 45, 90, 135, 180, 225, 360, 1440, 'SD1', 'SD2', 'SD3', 'SD4', 'SD5', 'SD6', 'SD7', 'SD8']
```

```python
# Extract time points
time_points = [15, 45, 90, 135, 180, 225, 360, 1440]
# Define marker styles
marker_styles = ['o', 's', '^', 'v', 'D', '*', 'x', '+', '<', '>']
# Define specified colors
colors = ['#800000','#f2b56f', '#f2a7da',  '#84c3b7','#71b7ed',  '#00008b', '#a020f0', '#1e90ff', '#ffdead', '#000000']
```

```python
# Prepare the plot
plt.figure(figsize=(12, 8))

# Loop through each sample to plot the data with error bars and different colors
for i in range(0, 10):
    conversion_rates = df.iloc[i, 1:9].astype(float)
    std_dev = df.iloc[i, 9:17].astype(float)  # Ensure this range matches the length of time_points
    
    # Plot error bars without changing their marker style
    plt.errorbar(time_points, conversion_rates, yerr=std_dev, linestyle='-', capsize=3, color=colors[i])
    
    # Plot data points with different markers and colors
    plt.plot(time_points, conversion_rates, marker=marker_styles[i], linestyle='-', color=colors[i], markersize=6, label=f'{df.iloc[i, 0]}')

# Customize the plot with bold Arial font
plt.xlabel('Time (minutes)', fontsize=14, fontweight='bold', fontname='Arial')
plt.ylabel('Conversion Rate (%)', fontsize=14, fontweight='bold', fontname='Arial')
plt.title('Conversion Rate (%) of NADPH Over Time in Total Turnover Assay', fontsize=16, fontweight='bold', fontname='Arial')

plt.legend(bbox_to_anchor=(1.05, 0.5), loc='center left')

# Set y-axis grid only
plt.grid(True, axis='y')

# Set y-axis limits
plt.ylim(-20, 100)

# Save and show the plot with higher resolution
plot_path = '/Users/Zachary/Library/CloudStorage/OneDrive-UniversityofCambridge/Project/total_turnover_assay/TTA_whole_plot.png'  # Change to your desired path
plt.savefig(plot_path, dpi=300, bbox_inches='tight')  # Increase dpi for higher resolution
plt.show()
```
