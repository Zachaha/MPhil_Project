```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data from the CSV file
file_path = "/Users/Zachary/Library/CloudStorage/OneDrive-UniversityofCambridge/Project/Focused_Library/2nd_3rd_data_for_figure.csv"
df = pd.read_csv(file_path)

# Extract the data for plotting
sorted_library_data = df['Sorted Library'].dropna().values
selected_mutant_data = df['Selected Mutants'].dropna().values

# Function to add jitter
def add_jitter(values, jitter_amount=0.2):
    return values + np.random.uniform(-jitter_amount, jitter_amount, size=values.shape)

# Calculate means and standard errors
mean_sorted_library = np.mean(sorted_library_data)
mean_selected_mutants = np.mean(selected_mutant_data)
sem_sorted_library = np.std(sorted_library_data) / np.sqrt(len(sorted_library_data))
sem_selected_mutants = np.std(selected_mutant_data) / np.sqrt(len(selected_mutant_data))

# Set font to Arial
plt.rcParams['font.family'] = 'Arial'

# Plotting
fig, ax = plt.subplots()

# Adjusted positions for the scatter plots
positions = [0, 1, 2, 3]

# Adding scatter plots with specified colors and jitter
ax.scatter(add_jitter(np.full_like(sorted_library_data, positions[1])), sorted_library_data, color='#4E6973', alpha=0.2, label='Sorted Library')
ax.scatter(add_jitter(np.full_like(selected_mutant_data, positions[2])), selected_mutant_data, color='#B6562E', alpha=0.4, label='Selected Hits')

# Adding dummy scatter plots to adjust spacing
ax.scatter([], [], color='none', alpha=0.0)  # Group 1 (Empty)
ax.scatter([], [], color='none', alpha=0.0)  # Group 2 (Empty)

# Adding mean lines and error bars
ax.hlines(mean_sorted_library, positions[1] - 0.26, positions[1] + 0.26, color='black', linestyle='-', linewidth=2)
ax.hlines(mean_selected_mutants, positions[2] - 0.26, positions[2] + 0.26, color='black', linestyle='-', linewidth=2)

# Adding error bars
ax.errorbar(positions[1], mean_sorted_library, yerr=sem_sorted_library, fmt='none', ecolor='black', capsize=5)
ax.errorbar(positions[2], mean_selected_mutants, yerr=sem_selected_mutants, fmt='none', ecolor='black', capsize=5)

# Labeling mean values
ax.text(positions[1] + 0.3, mean_sorted_library, f'{mean_sorted_library:.2f}', color='#4E6973', va='center', ha='left')
ax.text(positions[2] + 0.3, mean_selected_mutants, f'{mean_selected_mutants:.2f}', color='#B6562E', va='center', ha='left')

# Setting labels and titles
ax.set_xticks(positions)
ax.set_xticklabels(['', 'Sorted Library', 'Selected Hits', ''])
ax.set_ylabel('Fold-change over Wildtype')
ax.set_title('Fold-change Lysate Activity over WT')

# Removing the upper and right boundaries
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adding the legend
ax.legend()
# Save the figure
output_path = "/Users/Zachary/Library/CloudStorage/OneDrive-UniversityofCambridge/Project/Focused_Library/2nd_3rd_figure.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import textwrap

# Load the data from the CSV file
file_path = "/Users/Zachary/Library/CloudStorage/OneDrive-UniversityofCambridge/Project/Focused_Library/2nd_3rd_data_for_figure.csv"
df = pd.read_csv(file_path)

# Extract the data for plotting
sorted_library_data = df['Sorted Library'].dropna().values
selected_mutant_data = df['Selected Mutants'].dropna().values
selected_mutants_normalized_data = df['Selected Mutants Normalised by Concentration'].dropna().values

# Function to add jitter
def add_jitter(values, jitter_amount=0.15):
    return values + np.random.uniform(-jitter_amount, jitter_amount, size=values.shape)

# Calculate means and standard errors
mean_sorted_library = np.mean(sorted_library_data)
mean_selected_mutants = np.mean(selected_mutant_data)
mean_selected_mutants_normalized = np.mean(selected_mutants_normalized_data)

sem_sorted_library = np.std(sorted_library_data) / np.sqrt(len(sorted_library_data))
sem_selected_mutants = np.std(selected_mutant_data) / np.sqrt(len(selected_mutant_data))
sem_selected_mutants_normalized = np.std(selected_mutants_normalized_data) / np.sqrt(len(selected_mutants_normalized_data))

# Set font to Arial
plt.rcParams['font.family'] = 'Arial'

# Plotting
fig, ax = plt.subplots()

# Adjusted positions for the scatter plots
positions = [0, 1, 2, 3, 4]

# Adding scatter plots with specified colors and jitter
ax.scatter(add_jitter(np.full_like(sorted_library_data, positions[1])), sorted_library_data, color='#4E6973', alpha=0.2)#, label='Sorted Library')
ax.scatter(add_jitter(np.full_like(selected_mutant_data, positions[2])), selected_mutant_data, color='#B6562E', alpha=0.4)#, label='Selected Hits')
ax.scatter(add_jitter(np.full_like(selected_mutants_normalized_data, positions[3])), selected_mutants_normalized_data, color='#D98473', alpha=0.4)#, label='Selected Mutants Normalised by Concentration')

# Adding dummy scatter plots to adjust spacing
ax.scatter([], [], color='none', alpha=0.0)  # Group 1 (Empty)
ax.scatter([], [], color='none', alpha=0.0)  # Group 2 (Empty)

# Adding mean lines and error bars
ax.hlines(mean_sorted_library, positions[1] - 0.2, positions[1] + 0.2, color='black', linestyle='-', linewidth=2)
ax.hlines(mean_selected_mutants, positions[2] - 0.2, positions[2] + 0.2, color='black', linestyle='-', linewidth=2)
ax.hlines(mean_selected_mutants_normalized, positions[3] - 0.2, positions[3] + 0.2, color='black', linestyle='-', linewidth=2)

# Adding error bars
ax.errorbar(positions[1], mean_sorted_library, yerr=sem_sorted_library, fmt='none', ecolor='black', capsize=5)
ax.errorbar(positions[2], mean_selected_mutants, yerr=sem_selected_mutants, fmt='none', ecolor='black', capsize=5)
ax.errorbar(positions[3], mean_selected_mutants_normalized, yerr=sem_selected_mutants_normalized, fmt='none', ecolor='black', capsize=5)

# Labeling mean values
ax.text(positions[1] + 0.3, mean_sorted_library, f'{mean_sorted_library:.2f}', color='#4E6973', va='center', ha='left')
ax.text(positions[2] + 0.3, mean_selected_mutants, f'{mean_selected_mutants:.2f}', color='#B6562E', va='center', ha='left')
ax.text(positions[3] + 0.3, mean_selected_mutants_normalized, f'{mean_selected_mutants_normalized:.2f}', color='#D98473', va='center', ha='left')


# Setting labels and titles
ax.set_xticks(positions)
wrapped_labels = ['', 'Sorted Library', 'Selected Hits', '\n'.join(textwrap.wrap('Selected Mutants Normalised by Concentration', 20)), '']
ax.set_xticklabels(wrapped_labels)

ax.set_ylabel('Fold-change over Wildtype')
ax.set_title('Fold-change Lysate Activity over WT')

# Removing the upper and right boundaries
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adding the legend
#ax.legend()


# Save the figure
output_path = "/Users/Zachary/Library/CloudStorage/OneDrive-UniversityofCambridge/Project/Focused_Library/3_column_result.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')

plt.show()
```
