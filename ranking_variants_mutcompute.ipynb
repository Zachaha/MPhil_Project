```python
import pandas as pd
```

```python
all_mutation = pd.read_csv ("Mutcompute_5OCM.tsv", index_col=0)
```

```python
all_mutation.head()
```

```python
all_mutation.columns
```

```python
# Filter rows for each chain ID and create separate DataFrames
chain_A = all_mutation[all_mutation['chain_id'] == 'A']
chain_B = all_mutation[all_mutation['chain_id'] == 'B']
chain_C = all_mutation[all_mutation['chain_id'] == 'C']
chain_D = all_mutation[all_mutation['chain_id'] == 'D']
chain_E = all_mutation[all_mutation['chain_id'] == 'E']
chain_F = all_mutation[all_mutation['chain_id'] == 'F']
```

```python
print (chain_A.shape,chain_B.shape,chain_C.shape,chain_D.shape,chain_E.shape,chain_F.shape)
```

```python
average_log_ratiosA = chain_A.groupby(['position','wtAA', 'prAA']).agg({'avg_log_ratio': 'sum'}).sort_values(by='avg_log_ratio', ascending=False)#.reset_index(drop=True)
average_log_ratiosA.head(20)
```

```python
average_log_ratiosF = chain_F.groupby(['position','wtAA', 'prAA']).agg({'avg_log_ratio': 'sum'}).sort_values(by='avg_log_ratio', ascending=False)#.reset_index(drop=True)
average_log_ratiosF
```

```python
average_log_ratiosB = chain_B.groupby(['position','wtAA', 'prAA']).agg({'avg_log_ratio': 'sum'}).sort_values(by='avg_log_ratio', ascending=False)#.reset_index(drop=True)
average_log_ratiosB
```

```python
average_log_ratiosC = chain_C.groupby(['position','wtAA', 'prAA']).agg({'avg_log_ratio': 'sum'}).sort_values(by='avg_log_ratio', ascending=False)#.reset_index(drop=True)
average_log_ratiosC
```

```python
# Group by 'wtAA' and 'prAA' and calculate the average 'avg_log_ratio' across all chains
average_log_ratios = all_mutation.groupby(['position','wtAA', 'prAA']).agg({'avg_log_ratio': 'sum'}).reset_index()
average_log_ratios['avg_log_ratio'] /= 6

print(average_log_ratios)
```

```python
# Sort the DataFrame in descending order by 'avg_log_ratio'
average_log_ratios_sorted = average_log_ratios.sort_values(by='avg_log_ratio', ascending=False).reset_index(drop=True)
average_log_ratios_sorted.index = average_log_ratios_sorted.index + 1

print(average_log_ratios_sorted)
```

```python
average_log_ratios_sorted.to_csv('ranked_mutations.csv')
```
